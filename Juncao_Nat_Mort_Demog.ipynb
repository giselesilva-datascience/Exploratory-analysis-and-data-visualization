#### Fazer a junção dos 3 datasets em um só, para facilitar a análise.

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
v1=pd.read_csv('ETL_Natalidade_1900_a_2023.csv',sep=';')
v1.head()
```

```python
v2=pd.read_csv('ETL_Taxa_Mortalidade_1900_a_2023.csv',sep=';')
v2.head()
```

```python
v3=pd.read_csv('ETL_Indice_Demografico_1900_a_2023.csv',sep=';')
v3.head()
```

```python
comb_vars = v1.merge(v2, on='Anos', how='inner').merge(v3, on='Anos', how='inner')
comb_vars.head()
```

#### Renomear Colunas 

```python
Renomear = comb_vars.rename(columns={'Total de Nados-vivos': 'Nascimentos', 'Masculino_x': 'Nasc. M', 'Feminino_x': 'Nasc. F', 'Total de Obitos': 'Obitos', 
                                        'Masculino_y': 'Obitos M', 'Feminino_y': 'Obitos F', 'População Residente Total': 'Popupação'})
Renomear.head()
```

```python
Apagar_col = Renomear.drop(columns=['Nascimento Fora do Casamento','Taxa bruta de natalidade', 'Índice sintético de fecundidade', 'Obitos menos de 1 ano','Idade média da mulher ao nascimento do primeiro filho','Idade média da mulher ao nascimento de um filho','Taxa bruta de mortalidade', 'Taxa de mortalidade infantil', 'Taxa de mortalidade infantil','Esperanca de vida a nascenca','Taxa de crescimento natural','Taxa de crescimento migratório', 'Taxa de crescimento efetivo','Envelhecimento Populacional Total'])
Apagar_col.head()
```

```python
ETL_Dados_Parcial = Apagar_col.to_csv('ETL_Dados_Parcial.csv', sep=';', index=False)
```
